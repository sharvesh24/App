```python
from google_play_scraper import app, reviews_all
import pandas as pd
import spacy
from google.colab import files
from langdetect import detect
```

```python
id = 'com.microsoft.appmanager'
info = app(id)
print(info)
reviews = reviews_all(
    id,
    lang='en',
    country='in'
)
```

```python
num_reviews = 10000
limited_reviews = reviews[:num_reviews]
```

```python
df = pd.DataFrame(limited_reviews)
nlp = spacy.load('en_core_web_sm')
```

```python
def preprocess_review(review):
    review = ''.join(e for e in review if e.isalnum() or e.isspace())
    doc = nlp(review)
    tokens = []
    for token in doc:
        if not token.is_stop and not token.is_punct:
            tokens.append(token.lemma_.lower())
    cleaned_review = ' '.join(tokens)
    return cleaned_review
```

```python
def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False
```

```python
df = df[df['content'].apply(is_english)]
df['cleaned_content'] = df['content'].apply(preprocess_review)
```

```python
csv_filename = 'cleaned_play_store_reviews.csv'
df.to_csv(csv_filename, index=False)
```
